```python
import matplotlib.pyplot as plt
```

```python
from pathlib import Path

outdir = Path("out")
outdir.mkdir(exist_ok=True)
```

```python
def plot_linpack():
    labels = [
        "Threadripper PRO 5995WX 64C/128T @ 2.7GHz",
        "R7 7950X 16C/32T @ 4.5GHz",
        "R9 9950X 16C/32T @ 4.3GHz",
    ]
    tflops = [1.856, 0.972, 1.864]
    colors = ["midnightblue", "midnightblue", "lightseagreen"]
    fig, ax = plt.subplots(figsize=(12, 4))
    ax.set_title("HPL Linpack Benchmark", fontsize=16)
    ax.set_xlabel("TFLOP/s (Higher is better)")

    hbars = ax.barh(labels, tflops, color=colors)
    ax.bar_label(hbars, tflops)

    fig.savefig(outdir / "linpack.png", dpi=600, bbox_inches="tight")

    plt.show()


plot_linpack()
```

```python
def plot_stream():
    # I will not put this figure as the numbers from internet
    # Are Peak Theorical Bandwidth
    
    labels = ["Apple M4", "R9 9950X 16C/32T @ 4.3GHz"]
    gbs = [120, 69.673]
    colors = ["midnightblue", "lightseagreen"]
    fig, ax = plt.subplots(figsize=(12, 4))
    ax.set_title("STREAM Benchmark", fontsize=16)
    ax.set_xlabel("GB/s (Higher is better)")

    hbars = ax.barh(labels, gbs, color=colors)
    ax.bar_label(hbars, gbs)

    fig.savefig(outdir / "stream.png", dpi=600, bbox_inches="tight")

    plt.show()


plot_stream()
```
